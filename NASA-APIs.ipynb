```python
import urllib.request
from PIL import Image
import requests
from io import BytesIO
import codecs
import json
import requests
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#contents = urllib.request.urlopen("https://api.nasa.gov/planetary/apod?api_key=DEMO_KEY").read()

endpoint = 'https://api.nasa.gov/planetary/apod?'
api_key = 'DEMO_KEY'

nav_request = endpoint+'api_key={}'.format(api_key)
contents = requests.get(nav_request).json()
contents['url']
```

```python
img_url = contents['url']

response = requests.get(img_url)
img = Image.open(BytesIO(response.content))
```

```python
size = 600, 800
img.resize(size,resample=Image.BILINEAR)
```

```python
histo = img.histogram()
```

```python
plt.plot(histo)
```

```python
df = pd.DataFrame(histo)
```
